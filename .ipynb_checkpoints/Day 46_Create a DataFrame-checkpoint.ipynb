# Day 46: Create a DataFrame
Create a Dataframe using pandas. You are going to create a code to put the following into a Dataframe. You will use the
information in the table below. Basically, you are going to recreate this table using pandas. Use the information in the table to recreate the table.
year Title genre
2009 Brothers Drama
2002 Spider-Man Sci-fi
2009 WatchMen Drama
2010 Inception Sci-fi
2009 Avatar Fantasy

```python
import pandas as pd
```

```python
a = [2009, 'Brothers', 'Drama']
b = [2002, 'Spider-Man', 'Sci-fi']
c = [2009, 'WatchMen', 'Drama']
d = [2010, 'Inception', 'Sci-fi']
e = [2009, 'Avatar', 'Fantasy']
```

```python
df = pd.DataFrame(zip(a,b,c,d,e), index= ['year', 'Title', 'genre']).T
df
```
